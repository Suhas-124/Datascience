# 🌍 Introduction to Git & GitHub

## 🔧 What is Git?
Git is a **version control system** that lets you track changes in your code, collaborate with others, and easily manage multiple versions of your project.

Think of Git like a super-smart **Time Machine** for your code. You can:
- ⏪ Go back to previous versions
- 🔄 Work on multiple versions simultaneously (branches)
- ✅ Merge everyone's work together



## ☁️ What is GitHub?
GitHub is a **cloud-based hosting service** for Git repositories. It provides:
- A platform to **store** your Git repositories
- Tools for **collaboration**, **issue tracking**, and **code review**

🌐 In short: Git is the tool; GitHub is where you use it with others online.



# 🧰 Basic Git Commands (with Examples)

## 1️⃣ `git init`
**Initialize a Git repository.**
```bash
git init
```
🎯 This sets up the current directory as a Git repository and starts version tracking.



## 2️⃣ `git status`
**Check the current state of your repo.**
```bash
git status
```
🔍 Shows:
- Files that are tracked
- Files that are modified
- Files that are untracked (new files)



## 3️⃣ `git add <filename>` or `git add .`
**Stage your changes before committing.**
```bash
git add my_script.py
# or stage everything
git add .
```
🧺 Think of this as putting files in a basket to prepare for saving.



## 4️⃣ `git commit -m "Your message here"`
**Commit staged changes with a message.**
```bash
git commit -m "Initial commit with feature X"
```
📝 This saves your staged changes to the project history.

💡 Tip: Use clear messages so your future self understands what you did!



## 5️⃣ `.gitignore` file
**Tell Git what to ignore.**
Create a file called `.gitignore` and add patterns:
```plaintext
*.log
__pycache__/
.env
*.h5
*.csv
```
🚫 This prevents clutter and avoids pushing unnecessary or sensitive files.



# 🤝 Working with GitHub

## 6️⃣ Connect to GitHub
```bash
git remote add origin https://github.com/your-username/your-repo.git
```

## 7️⃣ Push your code
```bash
git push -u origin main
```
🌟 Your project is now on GitHub!

## 🌟 Let's Talk Identity in Git!

Git needs to know **who you are** when you make commits — so it can tag your name and email in each commit history.



### ✅ **Set Username and Email Globally**

This applies to **all your Git projects** on your system.

```bash
git config --global user.name "Your Full Name"
git config --global user.email "youremail@example.com"
```

🧩 Example:

```bash
git config --global user.name "Suhas S"
git config --global user.email "suhas@example.com"
```

📍These settings are saved in:

```bash
~/.gitconfig
```

You can view your global config using:

```bash
git config --global --list
```



### 🗂️ **Set Username and Email Locally (Per Project)**

You might want to use a different identity for **work** and **personal** projects.

🔸 First, navigate to your project folder:

```bash
cd ~/Desktop/office/project-folder/
```

🔸 Then, set user and email **locally** for that specific repo:

```bash
git config user.name "Work Name"
git config user.email "work@example.com"
```

💡 This stores identity settings in `.git/config` (inside that repo folder).

## 🎨 Recap Table

| Scope      | Command Example                                                   | Stored In                      |

| Global     | `git config --global user.name "Suhas"`                           | `~/.gitconfig`                 |
| Local Repo | `git config user.name "Office Git"` in that repo folder          | `.git/config` inside the repo  |



## 🧪 To Verify Current Settings


To explicitly see all:


## 🧾 **Seeing Commits in Git – The Detective Tools 🕵️‍♂️**

> Once you’ve committed code, these commands help you **view and analyze** your commit history!



### 🔍 1. `git log`  
Shows a detailed **history of all commits** in reverse order (latest first).

🧱 **Output includes:**  
- Commit hash (SHA)
- Author
- Date
- Commit message

🔧 **Example:**

```bash
git log
```

📌 Great when you want the **full story** of every commit.



### 📄 2. `git log --oneline`  
Gives a **summary** — each commit in a single line.

🔧 **Example:**

```bash
git log --oneline
```

🎯 Use this when you just want a **quick glance** at your history.

🧪 **Sample Output:**

```
9a8e9e2 Updated README file
c3d4b19 Fixed login bug
a0b1c3e Initial commit
```



### 📊 3. `git log --stat`  
Shows **which files changed**, and how many lines were added or removed.

🔧 **Example:**

```bash
git log --stat
```

🧩 Very helpful for **tracking code churn** (who changed what and how much).



### 🧬 4. `git log -p`  
Displays the **actual code differences (diffs)** introduced in each commit.

🔧 **Example:**

```bash
git log -p
```

📘 This is great for **reviewing code** before pushing or merging.



### 🧠 5. `git show <sha>`  
See **detailed info about one specific commit**, including changes.

🔧 **Example:**

```bash
git show 9a8e9e2
```

💡 Tip: You only need the **first few characters** of the SHA (usually 6+ is enough).



### 🧪 6. `git diff`  
Shows the difference between:
- Working directory 🧑‍💻 and
- Staging area (index) 🧷

🔧 **Example:**

```bash
git diff
```

If you've used `git add`, you can do:

```bash
git diff --staged
```

📌 Helps verify what changes are **yet to be committed**.

## 🧠 Summary Table

| Command | Purpose |
|--------|---------|
| `git log` | Full history of commits |
| `git log --oneline` | Short, 1-line summary per commit |
| `git log --stat` | File-level changes summary |
| `git log -p` | Code diffs introduced in each commit |
| `git show <sha>` | Detailed info for a specific commit |
| `git diff` | Changes not yet staged or committed |

````python
git log
git log --oneline
git log --stat
git log -p
git show <sha>
git diff
````

## 🌿 **Branching and Merging in Git**  
> Imagine branches as **parallel universes** of your project — perfect for experimenting without messing up the main timeline!



### 🌱 1. `git branch <name>` or `git branch <name> <sha-id>`  
Creates a **new branch** from the current HEAD, or from a specific commit using SHA ID.

🔧 **Examples:**

```bash
git branch feature-login           # From current HEAD
git branch bugfix-123 abcde12      # From specific commit
```



### 📜 2. `git branch`  
Lists **all branches**, highlighting the currently active one with `*`.

🔧 **Example:**

```bash
git branch
```

🧠 Output:
```
* main
  feature-login
  bugfix-123
```



### 🔀 3. `git checkout <branch>`  
Switches to the specified branch — like **jumping to another timeline** 🌌.

🔧 **Example:**

```bash
git checkout feature-login
```

🧠 Tip: Use `-b` to create and switch in one go:

```bash
git checkout -b new-feature
```



### 🕸️ 4. `git log --oneline --all`  
Shows a **compact list of commits** across **all branches**.

🔧 **Example:**

```bash
git log --oneline --all
```

🔎 Great for visualizing how branches have diverged and merged.



### 📊 5. `git log --oneline --all --graph`  
Displays a **commit tree graph** showing all branches and their relationships.

🔧 **Example:**

```bash
git log --oneline --all --graph
```

🧠 Looks something like:
```
* abc1234 (feature) Add login form
| * def5678 (main) Fix typo in README
|/
* fedcba9 Initial commit
```



### 🧹 6. `git branch -d <branch>` / `git branch -D <branch>`  
Deletes a branch locally.  
- `-d` (safe): Deletes only if fully merged  
- `-D` (force): Deletes regardless of merge

🔧 **Examples:**

```bash
git branch -d feature-login     # Safe delete
git branch -D old-branch        # Force delete
```



### 🧬 7. `git merge <branch>`  
Combines the **target branch’s work** into the **current branch** (usually `main`).

🔧 **Example:**

```bash
git checkout main
git merge feature-login
```

💡 This brings in all commits from `feature-login` to `main`.

## 🧠 Summary Table

| Command | Purpose |
|--------|---------|
| `git branch <name>` | Create a new branch |
| `git branch` | List all branches |
| `git checkout <branch>` | Switch to a branch |
| `git log --oneline --all` | See all commits from all branches |
| `git log --oneline --all --graph` | See visual commit graph |
| `git branch -d / -D <branch>` | Delete a branch (safe/force) |
| `git merge <branch>` | Merge a branch into the current branch |


### 🌈 Bonus Tips
- 🔄 You can also use `git switch` and `git restore` in newer Git versions.
- 📌 Try visual Git tools like **GitKraken** or `tig` (terminal UI) to explore branches visually.

## 🌐 WORKING WITH REMOTE REPO

Remote repositories are like your **project’s online backup** or **team collaboration center** — often hosted on platforms like **GitHub**, **GitLab**, or **Bitbucket**.



### ✅ 1. `git remote add origin <url>`  
This command **connects your local repo** to a **remote repository**, giving it a name (usually `origin`).

🔧 **Example:**

```bash
git remote add origin https://github.com/suhas/project.git
```

🔎 Now Git knows:  
Whenever you mention `origin`, it’s referring to `https://github.com/suhas/project.git`.

🧠 You can check if the remote was added using:

```bash
git remote -v
```



### 🚀 2. `git push origin master` or `git push origin main`  
This **uploads your local code** to the remote repo (e.g., GitHub) under the `master` or `main` branch.

🔧 **Example:**

```bash
git push origin main
```

📦 Your project is now available online for others to see, clone, or collaborate on.

🧠 If pushing for the first time and getting an error like:
```bash
fatal: The current branch main has no upstream branch.
```
Try this:
```bash
git push --set-upstream origin main
```



### 📥 3. `git pull origin master`  
This pulls the **latest changes from the remote** `master` branch to your local one — useful when others are also working on the same repo.

🔧 **Example:**

```bash
git pull origin master
```

📌 Think of this as syncing **downstream** from GitHub to your laptop.

🧠 If you're on `main`:
```bash
git pull origin main
```



### 📦 4. `git clone <url>`  
This **copies an entire remote repo** (including commit history) to your local machine.

🔧 **Example:**

```bash
git clone https://github.com/suhas/project.git
```

🎯 This creates a folder called `project/` on your system with the full codebase.

🧠 You can also rename the folder:

```bash
git clone https://github.com/suhas/project.git myfolder
```



## 🔁 Workflow Summary (Visual Order)

```
1. git clone <url>        ← Start with this if the repo already exists remotely
2. git remote add origin  ← Link your local repo to remote (only if it’s new)
3. git push               ← Push your local code to remote
4. git pull               ← Pull latest changes from remote
```



## 🧠 Pro Tip  
To remove/change remote:

```bash
git remote remove origin
git remote add origin <new-url>
```

---

````python
git remote add origin https://github.com/suhas/project.git
git remote -v
git push origin main
git pull origin main
git clone https://github.com/suhas/project.git
git remote remove origin
git remote add origin <new-url>
````
