## Bivariate/Multivariate

Analysis of two columns is bivariate.  
Analysis of more than two columns is multivariate.  


| Feature Relationship     | Plot Types                                |
|--------------------------|-------------------------------------------|
| Numerical-Numerical      | Scatter Plot, Line Plot, Hexbin Plot, Density Plot,KDE Plot |
| Numerical-Categorical    | Bar Plot, Box Plot, Violin Plot, Swarm Plot |
| Categorical-Categorical  | Heatmap, Cluster Map, Mosaic Plot, Grouped Bar Plot,Pair Plot(all details in one) |


## ðŸ’¡ **What is Bivariate Analysis?**
**Bivariate analysis** means analyzing the relationship between **two variables** to find patterns, correlations, or differences.

- **Example:** Is there a relationship between **age** and **salary**?  
- **Types of variables:**
  - **Numerical â†” Numerical** (e.g., Age vs. Salary)
  - **Categorical â†” Numerical** (e.g., Gender vs. Income)
  - **Categorical â†” Categorical** (e.g., Gender vs. Preferred Product)


## ðŸ§© **Steps to Perform Bivariate Analysis (General Procedure)**

### âœ… **Step 1: Understand the Data Types**
First, figure out whether the variables are **numerical** or **categorical**.

ðŸ‘‰ Example dataset:

| Age   | Salary   | Gender | Purchased |
|-------|----------|--------|-----------|
| 25    | 50000    | Male   | Yes       |
| 30    | 60000    | Female | No        |
| 28    | 52000    | Female | Yes       |

- **Age**: Numerical  
- **Salary**: Numerical  
- **Gender**: Categorical  
- **Purchased**: Categorical  

### âœ… **Step 2: Choose the Right Analysis Method Based on Variable Types**

| **Variable Type**       | **Recommended Methods**                             | **Visualization**          |
|-------------------------|-----------------------------------------------------|---------------------------|
| Numerical â†” Numerical    | Correlation, Regression, Scatter Plot              | Scatter Plot, Line Plot    |
| Categorical â†” Numerical  | Grouped Summary, Box Plot, Bar Chart               | Box Plot, Bar Chart        |
| Categorical â†” Categorical| Crosstab, Chi-Square Test, Heatmap                 | Heatmap, Stacked Bar Chart |



## ðŸ§ª **Bivariate Analysis Methods (with Code Examples)**



### ðŸ“Š **1. Numerical â†” Numerical: Use Correlation & Scatter Plot**

#### **Goal:** Check if two numerical variables are related.

âœ… **Steps:**
1. Calculate **correlation**.
2. Visualize with a **scatter plot**.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Age': [25, 30, 28, 32, 40],
    'Salary': [50000, 60000, 52000, 58000, 70000]
}
df = pd.DataFrame(data)

# 1. Correlation
correlation = df.corr()
print("Correlation Matrix:\n", correlation)

# 2. Scatter Plot
sns.scatterplot(data=df, x='Age', y='Salary')
plt.title('Scatter Plot: Age vs. Salary')
plt.show()
```

ðŸ“Œ **Interpretation:**  
- If the correlation is close to **1**: Strong positive relationship.  
- If the correlation is close to **-1**: Strong negative relationship.  
- If it's close to **0**: No relationship.



### ðŸ“¦ **2. Categorical â†” Numerical: Use Grouped Summary & Box Plot**

#### **Goal:** Compare numerical values across different categories.

âœ… **Steps:**
1. Use **groupby()** to get the summary.
2. Visualize with a **box plot** or **bar chart**.

```python
# Sample dataset
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'Salary': [50000, 60000, 52000, 58000, 70000]
}
df = pd.DataFrame(data)

# 1. Grouped Summary
grouped_summary = df.groupby('Gender')['Salary'].mean()
print("Grouped Summary:\n", grouped_summary)

# 2. Box Plot
sns.boxplot(data=df, x='Gender', y='Salary')
plt.title('Box Plot: Salary by Gender')
plt.show()
```

ðŸ“Œ **Interpretation:**  
- The box plot shows **distribution**.  
- Check if one category has significantly higher or lower values.


### ðŸ”¥ **3. Categorical â†” Categorical: Use Crosstab & Heatmap**

#### **Goal:** See how two categories relate.

âœ… **Steps:**
1. Use **crosstab()** to create a contingency table.
2. Visualize with a **heatmap**.

```python
# Sample dataset
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'Purchased': ['Yes', 'No', 'Yes', 'No', 'Yes']
}
df = pd.DataFrame(data)

# 1. Crosstab
crosstab = pd.crosstab(df['Gender'], df['Purchased'])
print("Crosstab:\n", crosstab)

# 2. Heatmap
sns.heatmap(crosstab, annot=True, cmap='Blues')
plt.title('Heatmap: Gender vs. Purchased')
plt.show()
```

ðŸ“Œ **Interpretation:**  
- The heatmap shows **frequency**.  
- You can see patterns like whether one gender purchases more frequently.


### ðŸ“ˆ **Bonus Tips for Effective Bivariate Analysis**
1. **Check for outliers** using box plots.
2. Use **log transformation** if the data is skewed.
3. Use **scatter plots** for quick insights and **correlation heatmaps** for a bigger picture.


## âœ… **Summary of Bivariate Analysis Steps**
| **Step**                 | **What to Do**                                     | **Example Tools**            |
|--------------------------|----------------------------------------------------|-----------------------------|

## Full code

````python
# 1. Numerical- Numerical

#scatter-plot bivariate
numerical_columns = ['total_bill','tip','size']

for col in numerical_columns:
    if col != 'tip':
        sns.scatterplot(x=df[col],y=df['tip'])
        plt.title(f"Scatter plot for {col} vs tip")
        plt.xlabel(col)
        plt.ylabel('tip')
        plt.show()
print('-'*50)


#Multivariate
for col in numerical_columns:
    if col != 'tip':
        sns.scatterplot(x=df[col],y=df['tip'],hue=df['sex'])
        plt.title(f"Multivariate Scatter plot for {col} vs tip")
        plt.xlabel(col)
        plt.ylabel('tip')
        plt.show()
````

````python
# Numerical - categorical

numerical_cat = ['tip']
categorical_cat = ['sex','smoker','day','time']

#bivariate
for cat in categorical_cat:
    sns.boxplot(x=df[cat],y=df['tip'])
    plt.title(f"Box Plot for {cat} vs tip")
    plt.xlabel(cat)
    plt.ylabel('tip')
    plt.show()

print('-'*50)

#Multivariate
for cat in categorical_cat:
    sns.boxplot(x=df[cat],y=df['tip'],hue=df['sex'])
    plt.title(f"Multivariate Box Plot for {cat} vs tip")
    plt.xlabel(cat)
    plt.ylabel('tip')
    plt.show()
````

````python
# categorical - categorical

#Bivariate
sns.pairplot(df)
print('-'*50)

#Multivariate
sns.pairplot(df,hue='sex')


# heatmap
corr = df.corr()
sns.heatmap(corr,annot=True,fmt='.2f')
````
