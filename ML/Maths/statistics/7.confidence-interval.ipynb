### **ğŸŒŸ How to Calculate a Confidence Interval:**

$$

- **$ \bar{x} $** = Sample mean (point estimate).

- **$ \sigma $** = Population standard deviation.
- **$ n $** = Sample size.
- **$ \frac{\sigma}{\sqrt{n}} $** = Standard error (SE), which measures how much the sample mean tends to vary from the true population mean.
