```python
import seaborn as sns
from scipy import stats

for col in X.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.histplot(X[col],kde=True)
    plt.title(col)

    plt.subplot(122)
    stats.probplot(X[col],dist='norm',plot=plt)
    plt.title(f"{col} pdf")
    plt.show()
```
