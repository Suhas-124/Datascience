```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/home/suhas/Desktop/office/Data Science/Notes/Data Analysis/Data/covid_toy.csv')
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_test.shape
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(
    transformers= [
        ('tnf1',SimpleImputer(),['fever']),
        ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
        ('tnf3',OneHotEncoder(sparse=False,drop='first'),['gender','city'])
        

    ],remainder='passthrough'
)
```

```python
s= transformer.fit_transform(X_train)
```

```python
d=transformer.transform(X_test)
```

```python
X_train_transformed = pd.DataFrame(s,)
X_test_transformed = pd.DataFrame(d)
```
