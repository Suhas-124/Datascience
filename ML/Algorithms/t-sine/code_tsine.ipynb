```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X
```

```python
y
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.scatter(X[:,0],X[:,1],c=y,cmap='viridis',s=60)
plt.title('Original Data (First two features)')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.colorbar()  # Show color bar to differentiate species
```

```python
tsne = TSNE(n_components=2,random_state=42)
X_tsne = tsne.fit_transform(X_scaled)
```

```python
X_tsne
```

```python
# Step 5: Visualize the results after t-SNE
plt.subplot(1, 2, 2)
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', s=60)
plt.title('t-SNE Visualization of Iris Dataset')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.colorbar(scatter)  # Show color bar to differentiate species
# Show the plot
plt.tight_layout()
plt.show()
```
