```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.datasets import load_iris
import seaborn as sns

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, 2]  # Select 'petal length' feature

# Convert feature to a TensorFlow tensor
X_tensor = tf.convert_to_tensor(X, dtype=tf.float32)

# Apply TensorFlow/Keras built-in activation functions
sigmoid_vals = tf.keras.activations.sigmoid(X_tensor).numpy()
tanh_vals = tf.keras.activations.tanh(X_tensor).numpy()
relu_vals = tf.keras.activations.relu(X_tensor).numpy()
leaky_relu_vals = tf.keras.activations.relu(X_tensor, alpha=0.01).numpy()  # Leaky ReLU
elu_vals = tf.keras.activations.elu(X_tensor).numpy()
softmax_vals = tf.keras.activations.softmax(tf.reshape(X_tensor, (-1, 1))).numpy()  # FIXED!

# Plot results
plt.figure(figsize=(12, 8))

activation_functions = [
    ('Sigmoid', sigmoid_vals),
    ('Tanh', tanh_vals),
    ('ReLU', relu_vals),
    ('Leaky ReLU', leaky_relu_vals),
    ('ELU', elu_vals),
    ('Softmax', softmax_vals[:, 0])  # Take first column for plotting
]

for i, (name, values) in enumerate(activation_functions):
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(x=X, y=values, label=name, color='b', edgecolor='black')
    plt.xlabel('Petal Length')
    plt.ylabel(f'{name}(x)')
    plt.title(name)
    plt.grid()

plt.tight_layout()
plt.show()
```
